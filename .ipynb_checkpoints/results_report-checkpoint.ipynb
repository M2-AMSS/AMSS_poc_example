### Load Python libraries

```python
import matplotlib.pyplot as plt
import my_stl_module  # module implemented 
import iplantuml  # Not necessary, used to render class diagram in jupyter

# explicitly select backend for rendering plot: 
# inline, nbagg (for interactive plot), TkAgg, Qt5Agg, WXAgg, and Agg
# %matplotlib inline
```

Module `my_stl_module` has been developped for handle STL ASCII file format. Three classes have been implemented : 
- Vetex - corner of triangle defined by with 3 coordinates,
- Facet - elementary triangular plane surface defined with 3 vertices,
- Solid - 3-D surface defined as the union of triangular facets joined by an elementary edge.

![Principle](img/stl_principle2.svg)

```python
%%plantuml

@startuml

Solid "n" - "n" Facet : contains >
Facet "n" - "3" Vertex : contains >

    
class Solid {
    str _name
    void add_facet()
    void add_facet_list()
    list get_all_points()
    list get_all_facets()
    list intersection_with_plan()
}

class Facet {
    Vertex v1azer
    Vertex v2
    Vertex v3
    list intersection_with_plan()
}

class Vertex {
    float x
    float y
    float z
}

@enduml  
```

### Load data from file

```python
chess_pawn = my_stl_module.Solid()  # create an empty object
chess_pawn.import_ascii("dataset/chess_pawn/chess_pawn_ascii.stl")
```

Display informations about `chess_pawn` object 

```python
print(chess_pawn)
```

### Set figure

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
(x_min, x_max, y_min, y_max, z_min, z_max) = chess_pawn.bounding_box()
ax.set_box_aspect((x_max - x_min, y_max - y_min, z_max - z_min))
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
```

### Plot the facets

```python
for facet in chess_pawn.get_all_facets():
    ax.plot(
        [facet.v1.x, facet.v2.x, facet.v3.x, facet.v1.x],
        [facet.v1.y, facet.v2.y, facet.v3.y, facet.v1.y],
        [facet.v1.z, facet.v2.z, facet.v3.z, facet.v1.z],
        color="steelblue",
        linewidth=0.25
    )

fig  # for display figure in the notebook
```

### Intersection with a plan

Equation of a plan: $ax + by + cz = d$ 

with
- $a = 7$
- $b = -1$
- $c = 2$
- $d = 16$

`chess_pawn.intersection_with_plan(a, b, c, d)` returns a list of line segments (line segment is a )

![Intersection](img/intersection.svg)

```python
a, b, c, d = 7.0, -1.0, 2.0, 16.0

line_segment_list = chess_pawn.intersection_with_plan(a, b, c, d)

# Display informations about intersection
print("Intersection gives", len(line_segment_list), "line segments")
```

### Plot line segments from intersection with plan

```python
for line_segment in line_segment_list:
    ax.plot(*line_segment, color="red")
    
fig  # for display figure in the notebook
```

### Zoom in on a detail
#### Define a region of interest

```python
x_min, x_max = +0.5, +3.5
y_min, y_max = -3.0, -0.0
z_min, z_max = +0.0, +5.0
```

#### Build the solid, intersection between the entire chess pawn solid and the region of interest

```python
chess_pawn_extract = chess_pawn.extract(
    x_min, x_max, y_min, y_max, z_min, z_max
)
print(chess_pawn_extract)
```

#### Set a new figure

```python
fig2 = plt.figure()
ax2 = fig2.add_subplot(projection='3d')
(x_min, x_max, y_min, y_max, z_min, z_max) = chess_pawn_extract.bounding_box()
ax2.set_box_aspect((x_max - x_min, y_max - y_min, z_max - z_min))
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_zlabel('z')
```

#### Draw all facets

```python
for facet in chess_pawn_extract.get_all_facets():
    ax2.plot(
        [facet.v1.x, facet.v2.x, facet.v3.x, facet.v1.x],
        [facet.v1.y, facet.v2.y, facet.v3.y, facet.v1.y],
        [facet.v1.z, facet.v2.z, facet.v3.z, facet.v1.z],
        color="steelblue",
        linewidth=0.5
    )
```

#### Build the solid, intersection between the chess pawn extract and the plan

```python
line_segment_list = chess_pawn_extract.intersection_with_plan(a, b, c, d)
print("Intersection gives", len(line_segment_list), "line segments")
```

#### Plot intersection (line segments and intersection points) and diplay figure

```python
for line_segments in line_segment_list:
    ax2.plot(*line_segments, color="red")

fig2
```
